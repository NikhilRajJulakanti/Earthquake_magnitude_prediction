```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data = pd.read_csv('C:/Users/DELL/Downloads/query.csv')

d = {}
for i in range(0, len(data['mag'])):
    if data.iloc[i]['time'][:4] not in d:
        d.update({data.iloc[i]['time'][:4]: data.iloc[i]['mag']})
    else:
        d[data.iloc[i]['time'][:4]] = max(d[data.iloc[i]['time'][:4]], data.iloc[i]['mag'])

```

```python
plt.plot(data['mag'],'b-')
plt.xlim(0, 5000)
plt.ylim(-10,20)
plt.show()
# magnitude distribution
import seaborn as sns
sns.histplot(data['mag'],bins=100,kde=True,color='blue')
plt.title('Magnitude Distribution')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()
```

```python
# rotate the x-axis labels
plt.bar(d.keys(), d.values(), color='violet')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Magnitude')
plt.title('Year vs Magnitude')
plt.show()
```

```python

india = {}
for i in range(len(data)):
    if 'India' in data.iloc[i]['place']:
        if data.iloc[i]['time'][:4] not in india:
            india.update({data.iloc[i]['time'][:4]: data.iloc[i]['mag']})
        else:
            india[data.iloc[i]['time'][:4]] = max(india[data.iloc[i]['time'][:4]], data.iloc[i]['mag'])
plt.bar(india.keys(), india.values(), color='green')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Magnitude')
plt.title('Year vs Magnitude in India')
plt.show()
```
